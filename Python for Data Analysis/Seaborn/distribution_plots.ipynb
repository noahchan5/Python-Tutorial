# Distribution Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

## Rug Plot
https://seaborn.pydata.org/generated/seaborn.rugplot.html#seaborn.rugplot

```python
# Creating a Rug Plot

sns.rugplot(x='body_mass_g', data=penguins)
```

```python
# Adjusting the height argument

sns.rugplot(x='body_mass_g', height=2, data=penguins)
```

```python
# Adding Rug Plots along x and y axis, changing figure size and setting y limit 

fig=plt.figure(figsize=(5,5))

ax = fig.add_axes([0,0,1,1])

ax.set_ylim(165,235)

sns.rugplot(x='body_mass_g',y='flipper_length_mm', 
            data=penguins)
```

```python
# Combining Rug Plot with Scatter Plot

fig=plt.figure(figsize=(5,5))
ax=fig.add_axes([0,0,1,1])

ax.set_ylim(165,235)

sns.rugplot(x='body_mass_g',y='flipper_length_mm', data=penguins)
sns.scatterplot(x='body_mass_g',y='flipper_length_mm', data=penguins)
```

## histplot
https://seaborn.pydata.org/generated/seaborn.histplot.html#seaborn.histplot

```python
# Creating a Hist Plot
sns.histplot(x='body_mass_g', data=penguins)
```

```python
# Adding a hue

sns.histplot(x='body_mass_g', data=penguins, hue='island')
```

```python
# Creating a 2d Hist Plot

sns.histplot(x='body_mass_g', data=penguins, y='island')
```

## KDE Plot
https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot

Kernel Density Estimation by Matthew Conlen: https://mathisonian.github.io/kde/

```python
# Creating a KDE Plot

sns.kdeplot(x='body_mass_g', data = penguins)
```

```python
# Combining with Rug Plot

sns.kdeplot(x='body_mass_g', data=penguins)
sns.rugplot(x='body_mass_g', data=penguins)
```

```python
# Creating a 2d KDE Plot

sns.kdeplot(x='body_mass_g', y = 'flipper_length_mm', data = penguins)
```

```python
# Combining with Scatter Plot

sns.kdeplot(x='body_mass_g',y='flipper_length_mm' ,data=penguins, fill=True)
sns.scatterplot(x='body_mass_g',y='flipper_length_mm' ,data=penguins)
```

## ecdfplot
https://seaborn.pydata.org/generated/seaborn.ecdfplot.html#seaborn.ecdfplot

```python
# Creating ECDF Plot on x axis variable

sns.ecdfplot(x='flipper_length_mm', data = penguins)
```

```python
# Creating ECDF Plot on y axis variable

sns.ecdfplot(y='flipper_length_mm', data = penguins)
```

```python
# Adding a hue

sns.ecdfplot(y='flipper_length_mm', data=penguins, hue='island')

plt.grid()
```

## displot
https://seaborn.pydata.org/generated/seaborn.displot.html

A displot can create distribution plots across a Facet Grid based on col and row variables

```python
sns.displot(x='flipper_length_mm', data=penguins, col='island', row='sex', kind='kde')
```
