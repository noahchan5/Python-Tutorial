# Regression Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

## corr()

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html

```python
# Using the corr() function to compute a pairwise correlation on the penguins DataFrame
penguins.corr()
```

# regplot
https://seaborn.pydata.org/generated/seaborn.regplot.html#seaborn.regplot

```python
# Creating a regplot

plt.figure(figsize=(10,10))

sns.regplot(x='flipper_length_mm', y ='bill_depth_mm', data=penguins)
```

```python
# Removing the confidence interval band

plt.figure(figsize=(10,10))

sns.regplot(x='flipper_length_mm', y='body_mass_g', data=penguins, ci=None)
```

# lmplot
https://seaborn.pydata.org/generated/seaborn.lmplot.html#seaborn.lmplot

```python
# lmplot can plot regplots on a FacetGrid

sns.lmplot(x='flipper_length_mm', y='body_mass_g', data=penguins, ci=None, height=5, aspect=1.5, row='island', col='sex')
```
