# Relational Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime as dt
```

```python
stocks = pd.read_excel('stocks.xlsx')
```

```python
stocks.head()
```

```python
stocks['Year']=stocks['Date'].dt.strftime('%Y')
```

```python
sns.set_theme()
```

# Line Plot
https://seaborn.pydata.org/generated/seaborn.lineplot.html

```python
# Creating a Line Plot on Seaborn

sns.lineplot(x='Date', y='Close', data=stocks, hue='Symbol')
```

```python
# Changing the figuresize, adding a hue and linestyle

plt.figure(figsize=(20,10))

sns.lineplot(x='Date', y='Close', data=stocks, hue='Symbol', linestyle='--')
```

```python
# Changing the palette to 'Set2'
# https://matplotlib.org/stable/tutorials/colors/colormaps.html

plt.figure(figsize=(20,10))

sns.lineplot(x='Date', y='Close', data=stocks, hue='Symbol', linestyle='--', palette='Set2')
```

```python
# Custom palette using hex codes

plt.figure(figsize=(20,10))

sns.lineplot(x='Date', y='Close', data=stocks, hue='Symbol', linestyle='--', palette=['#707070','#7cbb00', '#3b5998'])
```

```python
# Using subplots and plotting each stock symbol on seperate axes

fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(20,5))

sns.lineplot(x='Date', y='Close', data=stocks[stocks['Symbol']=='AAPL'],ax=ax1)
sns.lineplot(x='Date', y='Close', data=stocks[stocks['Symbol']=='FB'],ax=ax2)
sns.lineplot(x='Date', y='Close', data=stocks[stocks['Symbol']=='MSFT'],ax=ax3)
```

# relplot
https://seaborn.pydata.org/generated/seaborn.relplot.html

```python
# relplot allows you to create a Facet Grid based on categorical variables along the rows and columns

sns.relplot(x='Date', y='Close', data=stocks, hue='Symbol', kind='line')

plt.show()
```

```python
# Adding only a row variable and changing height and aspect

sns.relplot(x='Date', y='Close', data=stocks, kind='line', height=3, aspect=3.5, row='Symbol', hue='Symbol')

plt.show()
```

```python
# Adding only a column variable and changing height and aspect

sns.relplot(x='Date', y='Close', data=stocks, kind='line', height=5, aspect=1, col='Symbol', hue='Symbol')

plt.show()
```

# Scatter Plot
https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
sns.get_dataset_names()
```

```python
# loading sample dataset 'penguins' on to penguins variable
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
# Creating a Scatter Plot

fig = plt.figure(figsize=(10,10))

sns.scatterplot(x='body_mass_g', y='bill_length_mm', data=penguins)
```

```python
# Plotting a Scatter Plot along a Facet Grid using relplot by changing kind='scatter'

plt.figure(figsize=(10,10))

sns.relplot(x='body_mass_g', y='flipper_length_mm', data=penguins, row='sex',col='island', kind='scatter')
```
