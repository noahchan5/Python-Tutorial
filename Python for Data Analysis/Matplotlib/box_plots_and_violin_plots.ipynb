# Box Plots
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
players = pd.read_csv('players_21.csv')
```

```python
fig = plt.figure(figsize = (5,5))
ax1 = fig.add_axes([0,0,1,1])

ax1.boxplot(players['overall'], labels=['overall score'])

plt.show()
```

```python
# removing outliers via the showfliers argument

fig = plt.figure(figsize = (5,5))

ax1 = fig.add_axes([0,0,1,1])

ax1.boxplot(players['overall'], labels=['overall score'], showfliers=False)

plt.show()
```

```python
# setting vert to False

fig = plt.figure(figsize = (5,5))

ax1 = fig.add_axes([0,0,1,1])

ax1.boxplot(players['overall'], labels=['overall score'], showfliers=False, vert=False)

plt.show()
```

## Data Preperation to create two DataFrames for two seperate teams

```python
inter = players[players['club_name']=='Inter']['overall']
```

```python
dortmund = players[players['club_name']=='Borussia Dortmund']['overall']
```

```python
data = [inter, dortmund]
```

```python
type(data)
```

## Box Plots comparing two different teams

```python
fig = plt.figure(figsize = (5,5))
ax1 = fig.add_axes([0,0,1,1])

ax1.boxplot(data)

plt.show()
```

```python
# Adding labels 

fig = plt.figure(figsize = (5,5))

ax1 = fig.add_axes([0,0,1,1])

ax1.boxplot(data, labels = ['Inter Milan', 'Borussia Dortmund'])

plt.show()
```

# Violin Plots
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.violinplot.html

```python
# Creating Violin Plot +  adding xticks

fig = plt.figure(figsize = (5,5))
ax1 = fig.add_axes([0,0,1,1])

ax1.violinplot(data)

plt.xticks([1,2],['Inter', 'Dortmund'])

plt.show()
```

```python
# Adding Box Plot on to same axes as Violin Plot 

fig = plt.figure(figsize = (5,5))
ax1 = fig.add_axes([0,0,1,1])

ax1.violinplot(data)
ax1.boxplot(data)

plt.xticks([1,2],['Inter', 'Dortmund'])

plt.show()
```
