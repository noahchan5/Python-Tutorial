# Creating a Plot Area

```python
import matplotlib.pyplot as plt
```

# figure()

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html

```python
# creating a figure object using the figure() method, assigning it to the variable fig
fig = plt.figure()
```

```python
# As you can see we have a figure object but with 0 axes
fig
```

# add_axes()

https://matplotlib.org/stable/api/figure_api.html?highlight=add_axes#matplotlib.figure.Figure.add_axes

```python
# We can use the add_axes method to add axes to the figure object
fig = plt.figure().add_axes(rect=[0,0,1,1])
```

```python
# The 3rd and 4th elements in the rect argument determine the scale of the width and height
# The default width / height is 6.4 / 4.8
fig = plt.figure().add_axes(rect=[0,0,0.5,0.5])
```

```python
# Returns the height at a scale of 0.5
fig = plt.figure().add_axes(rect=[0,0,1,0.5])
```

```python
# The figsize argument in the figure() method will allow you to create an object with the specified width and height
fig = plt.figure(figsize=(10,10)).add_axes(rect=[0,0,1,1])
```

```python
# Width and Height of 10 x 10 at a scale of 0.5 returns an object of 5 x 5
fig = plt.figure(figsize=(10,10)).add_axes(rect=[0,0,0.5,0.5])
```

```python
# You can assign the axes object to a variable, you can then call methods on the axes
fig = plt.figure(figsize=(5,5))

axes = fig.add_axes(rect=[0,0,1,1])
```

```python
# Adding two axes objects, they overlap due to same left, bottom positions in the rect argument
fig = plt.figure(figsize=(5,5))

axes1 = fig.add_axes(rect=[0,0,1,1])
axes2 = fig.add_axes(rect=[0,0,1,1])
```

```python
# Axes are side by side as the rect argument position is at a scale of 1 x the width to the right
fig = plt.figure(figsize=(5,5))

axes1 = fig.add_axes(rect=[0,0,1,1])
axes2 = fig.add_axes(rect=[1,0,1,1])
```

```python
# rect = [1,1,x,x] means to the right by 5 inches and up by 5 inches
fig = plt.figure(figsize=(5,5))

axes1 = fig.add_axes(rect=[0,0,1,1])
axes2 = fig.add_axes(rect=[1,1,1,1])
```

```python
# rect = [1,-1,x,x] means to the right by 5 inches and down by 5 inches

fig = plt.figure(figsize=(5,5))

axes1 = fig.add_axes(rect=[0,0,1,1])
axes2 = fig.add_axes(rect=[1,-1,1,1])
```

```python

fig = plt.figure(figsize=(5,5))

axes1 = fig.add_axes(rect=[0,0,1,1])
axes2 = fig.add_axes(rect=[1.5,0,1,1])
```

```python
# rect = [0.1,0.1,0.5,0.5] means to the right by 0.5 inches and down by 0.5 inches, and scale the image down 50%

fig = plt.figure(figsize=(5,5))

axes1 = fig.add_axes(rect=[0,0,1,1])
axes2 = fig.add_axes(rect=[0.1,0.1,0.5,0.5])
```

# Adjusting the axis

set_xlim():  https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlim.html

set_ylim(): https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylim.html


```python
fig = plt.figure(figsize=(5,5))

axes1 = fig.add_axes(rect=[0,0,1,1])
axes2 = fig.add_axes(rect=[1.5,0,1,1])
```

```python
# Setting a range on the y axis of axes1 to 0 to 100
fig = plt.figure(figsize=(5,5))

axes1 = fig.add_axes(rect=[0,0,1,1])
axes2 = fig.add_axes(rect=[1.5,0,1,1])

axes1.set_ylim(0,100)
```

```python
# Setting a range on the y axis of axes1 to 0 to 100
# Setting a range on the y axis of axes2 to 0 to 50
fig = plt.figure(figsize=(5,5))

axes1 = fig.add_axes(rect=[0,0,1,1])
axes2 = fig.add_axes(rect=[1.5,0,1,1])

axes1.set_ylim(0,100)
axes2.set_xlim(0,50)
```

# twinx()

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.twinx.html

```python
# twinx() allows you to create a twin Axes sharing the xaxis.

fig = plt.figure(figsize=(5,5))
axes1 = fig.add_axes([0,0,1,1])
axes2 = axes1.twinx()
axes1.set_ylim(0,100)
axes2.set_ylim(0,10)
```

# Adding a Title and Labels

set_title(): https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html?highlight=set_title#matplotlib.axes.Axes.set_title

xlabel(): https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html

ylabel(): https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html

```python
# Adding a title to both axes
fig = plt.figure(figsize=(5,5))

axes1 = fig.add_axes(rect=[0,0,1,1])
axes2 = fig.add_axes(rect=[1.5,0,1,1])

axes1.set_title('Chart 1')
axes2.set_title('Chart 2')
```

```python
# ... adding an x and y label to both axes
fig = plt.figure(figsize=(5,5))

axes1 = fig.add_axes(rect=[0,0,1,1])
axes2 = fig.add_axes(rect=[1.5,0,1,1])

axes1.set_title('Chart 1')
axes2.set_title('Chart 2')

axes1.set_xlabel('X')
axes1.set_ylabel('Y')

axes2.set_xlabel('X')
axes2.set_ylabel('Y')
```
