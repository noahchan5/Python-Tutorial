# Style and Presentation

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
penguins=sns.load_dataset('penguins')
```

```python
penguins.dropna(inplace=True)
```

```python
x=penguins['body_mass_g']
y=penguins['flipper_length_mm']
```

## Title and Labels

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])

axes.set_title('Chart Title', fontsize=30)
axes.set_ylabel('Flipper Length (mm)', fontsize=20)
axes.set_xlabel('Body Mass (g)', fontsize=20)

sns.scatterplot(data=penguins,x=x,y=y)

plt.show()
```

```python
ax1=sns.scatterplot(data=penguins,x=x,y=y)

ax1.set(title='Chart1',xlabel='Body Mass (g)', ylabel='Flipper Length (mm)')

sns.set(font_scale=1.2)

plt.show()
```

## Figure Size

```python
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_axes([0,0,1,1])

sns.scatterplot(data=penguins,x=x,y=y)

ax1.set_title('Chart Title')
plt.show()
```

```python
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_axes([0,0,1,1])

sns.set_context('poster')
sns.scatterplot(data=penguins,x=x,y=y)

plt.show()
```

```python
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_axes([0,0,1,1])

sns.set_context('notebook')
sns.scatterplot(data=penguins,x=x,y=y)

plt.show()
```

## Despine
https://seaborn.pydata.org/generated/seaborn.despine.html

```python
sns.set_theme(style='ticks')
```

```python
sns.scatterplot(data=penguins,x=x,y=y)

plt.show()
```

```python
sns.scatterplot(data=penguins,x=x,y=y)
sns.despine()

plt.show()
```

```python
sns.scatterplot(data=penguins,x=x,y=y)
sns.despine(left=True, bottom=True)

plt.show()
```
