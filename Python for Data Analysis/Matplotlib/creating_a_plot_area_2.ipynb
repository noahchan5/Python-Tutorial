# Subplots

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

```python
# You can use subplots to create many plots, it takes a nrows and ncols argument to determine the subplots across each row and column
plt.subplots(1,1)
```

```python
# adding two subplots of 1 row and 2 columns
plt.subplots(1,2)
```

```python
# Adding a subplot of 2 rows and 2 columns
plt.subplots(2,2)
```

```python
# tight_layout=True will fix the layour
plt.subplots(2,2, tight_layout=True)
```

```python
# Assigning the subplot object to a variable x
x = plt.subplots(1,1)
```

```python
# ax will be an array storing two axes objects as elements
fig, ax = plt.subplots(1,2)
```

```python
# indexing to return the first axes object
ax[0]
```

```python
fig,ax = plt.subplots(1,2)
ax[0].set_title('first axes')
```

```python
# unpacking to assign the subplots to two seperate variables
fig, (ax1,ax2) = plt.subplots(1,2)
```

so: so now I've unpacked even more and ax1 is the first axes variable, and ax2 is a variable storing the second axes

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax2.set_title('ax2 title')
```

```python
# unpacking to assign the subplots to 4 seperate variables
fig, [[ax1,ax2],[ax3,ax4]] = plt.subplots(2,2)
```

```python
fig, [[ax1,ax2],[ax3,ax4]] = plt.subplots(2,2, tight_layout=True)
ax1.set_title('ax1')
ax2.set_title('ax2')
ax3.set_title('ax3')
ax4.set_title('ax4')
```
