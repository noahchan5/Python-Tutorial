# Scatter Plots

## Data Preparation

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
players = pd.read_csv('players_21.csv')
```

```python
# This setting allows you to view all columns
pd.set_option('display.max_columns', None)
```

```python
players.head()
```

```python
players_100 = players.head(100)
```

# Scatter Plot
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
fig = plt.figure(figsize = (10,10))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_title('Overall ability vs Wages')
ax1.set_xlabel('Overall ability')
ax1.set_ylabel('Wages')

ax1.scatter(x = players_100['overall'], y=players_100['wage_eur'])
```

```python
# adding the s argument for size

fig = plt.figure(figsize = (10,10))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_title('Overall ability vs Wages')
ax1.set_xlabel('Overall ability')
ax1.set_ylabel('Wages')

ax1.scatter(x = players_100['overall'], y=players_100['wage_eur'], s=30)
```

```python
# scaling down the size of the bubbles

fig = plt.figure(figsize = (5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_title('Overall ability vs Wages')
ax1.set_xlabel('Overall ability')
ax1.set_ylabel('Wages')

ax1.scatter(x = players_100['overall'], y=players_100['wage_eur'], s=players_100['value_eur']/150000)
```
