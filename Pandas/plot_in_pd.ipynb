# Plot in Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('air_quality_no2.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
type(df)
```

### DF Plot

```python
df.plot()
```

### Specific Column Plot

```python
df['station_paris'].plot()
```

### Scatter Plot

```python
df.plot.scatter(x='station_london', y = 'station_paris', alpha=0.5) # alpha = opiticity
```

```python
[method_name
    for method_name in dir(df.plot)
    if not method_name.startswith("_")]
```

### Box Plot

```python
df.plot.box()
```

### Each columns in a separate subplot

```python
axs = df.plot.area(figsize=(12,4), subplots=True)
# subplots=True >>> separate subplot
```

```python
fig, axs = plt.subplots(figsize=(12,4))

df.plot.area(ax = axs)
axs.set_ylabel('NO$_2$ concentration') # $_2_$ for lowercase

fig.savefig('no2_concentration.png')
```
