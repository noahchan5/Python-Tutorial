# Style and Presentation

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.arange(0,10,0.1)
```

```python
x.shape
```

```python
y1 = x
y2 = x**2
y3 = x**3
```

```python
(y1.shape,y2.shape,y3.shape)
```

## Labels and Font Size

#### Links and resources:
* https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html
* https://matplotlib.org/stable/gallery/pyplots/fig_axes_labels_simple.html
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yticks.html

```python
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_axes([0,0,1,1])

ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

ax1.set_xlim(0,10)
ax1.set_ylim(0,100)

ax1.set_title('Chart Title')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')

plt.show()
```

```python
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_axes([0,0,1,1])

ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

ax1.set_title('Chart Title Left', loc='left')
ax1.set_title('Chart Title', loc='center')
ax1.set_title('Chart Title Right', loc='right')
ax1.set_xlabel('x label', )
ax1.set_ylabel('y label')

plt.show()
```

```python
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_axes([0,0,1,1])

ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

ax1.set_title('Chart Title', fontsize=40, color='blue', fontname='Arial')
ax1.set_xlabel('x label', fontsize=20, fontname='Times New Roman', rotation='vertical')
ax1.set_ylabel('y label', fontsize=20, fontname='Helvetica')

plt.show()
```

```python
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_axes([0,0,1,1])

ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

ax1.set_title('Chart Title', fontsize=40, color='blue', fontname='Arial')
ax1.set_xlabel('x label', fontsize=20, fontname='Times New Roman', rotation='vertical')
ax1.set_ylabel('y label', fontsize=20, fontname='Helvetica')

plt.xticks([0,2,4,6,8,10],['zero','two','four','six','eight','ten'])

plt.show()
```

```python
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_axes([0,0,1,1])

ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

ax1.set_title('Chart Title', fontsize=40, color='blue', fontname='Arial')
ax1.set_xlabel('x label', fontsize=20, fontname='Times New Roman', rotation='vertical')
ax1.set_ylabel('y label', fontsize=20, fontname='Helvetica')

plt.xticks([0,2,4,6,8,10],['zero','two','four','six','eight','ten'], rotation='vertical')

plt.show()
```

## Legends

#### Links and resources:
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

```python
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_axes([0,0,1,1])

ax1.plot(x,y1, label='y=x')
ax1.plot(x,y2, label='y=x**2')
ax1.plot(x,y3, label='y=x**3')

plt.legend()

plt.show()
```

```python
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_axes([0,0,1,1])

ax1.plot(x,y1, label='y=x')
ax1.plot(x,y2, label='y=x**2')
ax1.plot(x,y3, label='y=x**3')

plt.legend(loc='center')

plt.show()
```

### Legends with categorical plots

```python
df = pd.read_csv('top_10_countries.csv')
```

```python
df.head(2)
```

```python
fig = plt.figure(figsize = (10,10))
ax1 = fig.add_axes([0,0,1,1])

for country in df['Country / Dependency']:
    ax1.bar('Country / Dependency', 'Population', data = df[df['Country / Dependency']==country], label=country)


plt.xticks(rotation='vertical')

plt.legend()

plt.show()
```

## Annotating

#### Links and resources:
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html
* https://matplotlib.org/stable/tutorials/text/annotations.html

```python
fig = plt.figure(figsize = (5,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

plt.annotate('green line',(2,200))
plt.arrow(3,190, dx=2,dy=-60)

plt.show()
```

```python
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

ax1.annotate('green line',
            xy=(7, 350), xycoords='data',
            xytext=(0.5, 0.5), textcoords='axes fraction',
            arrowprops=dict(facecolor='green', edgecolor='w'),
            horizontalalignment='left', verticalalignment='top')

plt.show()
```

## Grids

#### Links and resources:
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html
