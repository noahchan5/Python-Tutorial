# Categorical Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
countries_df = pd.read_csv('top_10_countries.csv')
```

```python
countries_df
```

## Bar Plots
https://seaborn.pydata.org/generated/seaborn.barplot.html

```python
# simple barplot with seaborn
sns.barplot(x = countries_df['Region'], y=countries_df['Population'])
```

```python
# Removing the confidence intervals

sns.barplot(x = countries_df['Region'], y=countries_df['Population'], ci=None)
```

```python
# Horizontal Orientation

sns.barplot(y = countries_df['Region'], x=countries_df['Population'], ci=None, orient="h")
```

```python
# Adding data argument to reference columns by name rather than indexing the DataFrame

sns.barplot(x = 'Region', y='Population', data=countries_df, ci=None)
```

```python
# Changing the estimator to sum, it is mean by default

sns.barplot(x = 'Region', y='Population', data=countries_df, ci=None, estimator=sum)
plt.show()
```

```python
# Adding hue

sns.barplot(x = 'Region', y='Population', data=countries_df, estimator=sum, ci=None, hue='Region', dodge=False)
plt.show()
```

```python
# Using matplotlib and seaborn concurrently can allow you to format your visual

plt.figure(figsize=(10,10))

sns.barplot(x = 'Country', y='Population', data=countries_df, estimator=sum, ci=None, hue='Region')

plt.legend(loc='center')

plt.show()
```

```python
# Setting dodge=False will prevent the bars from 'shifting'

plt.figure(figsize=(10,10))

sns.barplot(x = 'Country', y='Population', data=countries_df, estimator=sum, ci=None, hue='Region', dodge=False)

plt.legend(loc='center')

plt.show()
```

```python
# Using subplots to plot on 2 axes

fig, (ax1,ax2) = plt.subplots(figsize=(20,10),nrows=1,ncols=2, tight_layout=True)

sns.barplot(x=countries_df['Region'],y=countries_df['Population'], ci=None, estimator=sum, ax=ax1)
sns.barplot(x=countries_df['Country'],y=countries_df['Population'], ci=None, estimator=sum, ax=ax2)

plt.show()
```

# Count Plot
https://seaborn.pydata.org/generated/seaborn.countplot.html

```python
# Count Plot uses a count estimator and takes one axis argument
sns.countplot(x=countries_df['Region'])
```

# Box Plots and Violin Plots
https://seaborn.pydata.org/generated/seaborn.boxplot.html

https://seaborn.pydata.org/generated/seaborn.violinplot.html

```python
players=pd.read_csv('players_21_top_100.csv')
```

```python
pd.set_option('display.max_columns',False)
```

```python
players.head(1)
```

```python
# Creating a box plot

sns.boxplot(y=players['overall'])
```

```python
# Removing outliters by specifying fliersize=0

sns.boxplot(x=players['overall'], fliersize=0)
```

```python
# Adding an x and y axis argument 2 create a 2d box plot

sns.boxplot(y=players['overall'], x=players['league_name'])
```

```python
# Changing the figure size with plt

plt.figure(figsize=(10,10))
sns.boxplot(y=players['overall'], x=players['league_name'])
```

```python
# Violin Plot

plt.figure(figsize=(10,10))
sns.violinplot(y=players['overall'], x=players['league_name'])
```

# Strip Plot
https://seaborn.pydata.org/generated/seaborn.stripplot.html

```python
# Creating a Strip Plot

plt.figure(figsize=(10,10))
sns.stripplot(y=players['overall'], x=players['league_name'], color='black')
```

```python
# Strip Plot with Box Plot
plt.figure(figsize=(10,10))
sns.stripplot(y=players['overall'], x=players['league_name'], color='black')
sns.boxplot(y=players['overall'], x=players['league_name'])
```

# Swarm Plot
https://seaborn.pydata.org/generated/seaborn.swarmplot.html

```python
# Creating a Swarm Plot

plt.figure(figsize=(10,10))
sns.swarmplot(y=players['overall'], x=players['league_name'])
```

# Point Plot
https://seaborn.pydata.org/generated/seaborn.pointplot.html

```python
# 1d Point Plot

sns.pointplot(x = players['overall'])
```

```python
# 2d Point Plot

sns.pointplot(x = players['overall'], y=players['preferred_foot'])
```

```python
# Adding a hue

sns.pointplot(x=players['overall'], hue=players['preferred_foot'], y = players['league_name'])
```
